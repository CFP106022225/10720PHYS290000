```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.random(1000)
print(x)
```

```python
plt.hist(x,10)

plt.xlabel("x value")
plt.ylabel("Number of elements")
plt.show()
```

```python
mu,sigma = 1, 0.5 #mean and standard deviation
n = np.random.normal(mu,sigma,10)
plt.hist(n,10)

#plt.slim(-2.2)
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import math

h = 6.62*10**-34  #SI unit
c = 3*10**8  #speed of light
k = 1.38*10**-23  #Boltzmann Constant JK^-1

num = 500

v = np.linspace(10**0, 10**15.5, num, endpoint=True)
print(v)

T = 7000
B = 2*h*v**3/c**2/(math.e**(h*v/k/T)-1)
print(B)

#plt.errorbar(v, B, yerr=0.5*B, xerr=0.1*v, ecolor="green")

plt.plot(v, B)
#plt.xlim(10**15, 10**15.2)
#plt.ylim(10**-14, 10**-15)
#plt.plot(v, B, ".")
plt.xlabel("Frequency(Hz)")
plt.ylabel("Inensity Watts/Hz/m^2")

dB = np.random.normal(0., 10**-8.5, num)
B += dB

plt.plot(v,B,"c.")
#plt.loglog(v,B)

plt.show()
```
